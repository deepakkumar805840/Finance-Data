# EDA

# Gender

```python
sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for    xyz   in   ax   .    containers:
    ax.bar_label(xyz)
```

```python
df   .  groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen= df   .  groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
ay=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for xyz in ay   .     containers:
    ay.bar_label(xyz)
```

# From the above graph we can seethat most of the buyers are female. female(7.4 cr), male(3.19 cr)

# Age

```python
df.shape
```

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_age,x='Age Group', y='Amount')
```

```python
ax=sns.barplot(data=sales_age,x='Age Group', y='Amount')
for qw in ax.containers:
    ax.bar_label(qw)
```

#From the above graph we can see that most of the buyers are of age group 26-35

# State

```python
df.columns
```

```python
df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
labelbar= sns.barplot(data=sales_state,x='State',y='Amount')
for qw in labelbar.containers:
    labelbar.bar_label(qw)
```

```python
# from the above graph we can see the most orders are from UP, MH, KT
```

# Marital Status & Gender

```python
qw=sns.countplot(data=df, x='Marital_Status', hue='Gender')
sns.set(rc={'figure.figsize':(10,5)})
for we in qw.containers:
    qw.bar_label(we)
```

```python
df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
Sales_Marital_Status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=Sales_Marital_Status, x='Marital_Status',y='Amount',hue='Gender')
```

```python
qw=sns.barplot(data=Sales_Marital_Status, x='Marital_Status',y='Amount',hue='Gender')
for we in qw.containers:
    qw.bar_label(we)
```

```python
# from the above graph we can see that most of the buyers are married women.
```
